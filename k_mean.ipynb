```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

```

```python

# Step 1: Generate synthetic data
n_samples = 300
n_features = 2
n_clusters = 4
random_state = 42

data, true_labels = make_blobs(n_samples=n_samples, 
                               n_features=n_features, 
                               centers=4, 
                               cluster_std=1.0)
```

```python

# Step 2: Apply k-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
kmeans.fit(data)

# Retrieve the loss (inertia)
loss = kmeans.inertia_
print(f"Loss (Inertia): {loss}")
```

```python

# Step 3: Retrieve cluster labels and centroids
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 4: Visualize the clusters
plt.figure(figsize=(8, 6))
for i in range(n_clusters):
    plt.scatter(data[cluster_labels == i, 0], data[cluster_labels == i, 1], label=f"Cluster {i}")
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
